```python
support.groupby(['contributor_lastname', 'contributor_firstname']).amount.sum().reset_index().sort_values('amount', ascending = False)

# Group and sum contributions by contributor name to group contributors with more than one contribution in dataframe format sorted by contribution amount
# Reset index needs to go after you've pulled the data, after amount.sum
# Python reads from right to left, so order of methods is important
```
